```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('commodity.csv')
df = df[df['Symbol'] == 'US Wheat']
```

```python
#visulaize df by x=  Date, y = Open on plot 
df['Date'] = pd.to_datetime(df['Date'])
#Date >= 2015
df1 = df[df['Date'] >= '2015-01-01']
df1
```

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(40,7))
ax.plot(df1['Date'], df1['Open'])
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.scatter(df1['Date'], df1['Open'])
dateFmt = mdates.DateFormatter('%Y-%m-%d')

ax.xaxis.set_major_formatter(dateFmt)


plt.show()
```

```python
df2 = df[df['Date'] >= '2021-01-01']
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(40,7))
ax.plot(df2['Date'], df2['Open'])
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.scatter(df2['Date'], df2['Open'])
dateFmt = mdates.DateFormatter('%Y-%m-%d')

ax.xaxis.set_major_formatter(dateFmt)


plt.show()
```
